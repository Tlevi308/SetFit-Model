```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/drive/MyDrive/SetFit - 29.01.23/dataset.csv')
```

```python
df = df.iloc[:,1:]
```

```python
df = df.rename(columns={'description': 'text'})
df = df.rename(columns={'title': 'label'})
df = df[df.label != 1305]
df = df[df.label != 1306]
df = df[df.label != 1313]
df = df[df.label != 1326]
df = df[df.label != 1310]
df = df[df.label != 3211]
df = df[df.label != 3213]
df = df[df.label != 508]
```

```python
df['label'].value_counts(normalize=True)
```

```python
from sklearn.utils import shuffle
df = shuffle(df)
```

```python
from sklearn.model_selection import train_test_split
                                                      # stratify for fer split data - equal distribution
train_df, test_df = train_test_split(df,test_size=.2, stratify=df['label'])
```

```python
pip install setfit  
```

```python
from setfit import SetFitModel, SetFitTrainer
from datasets import Dataset
```

```python
train_ds = Dataset.from_pandas(train_df)
test_ds = Dataset.from_pandas(test_df)
```

```python
# Load SetFit model from Hub
model = SetFitModel.from_pretrained("sentence-transformers/paraphrase-mpnet-base-v2")

# Create trainer
trainer = SetFitTrainer(
    model=model,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    #loss_class=CosineSimilarityLoss,
    batch_size=16,
    num_iterations=20, # Number of text pairs to generate for contrastive learning
    num_epochs=1 # Number of epochs to use for contrastive learning
)
```

```python
trainer.train()
```

```python
 trainer.evaluate()
```
